# 3

```python
drinks = drinks_content[["Beverage_category", "Beverage", "Beverage_prep", "Calories"]].copy()
```

```python
drinks.isna().sum()
#no NAs
```

# 4

```python
drinks["Calorie_diff"] = 135 - drinks["Calories"]
```

# 5

```python
drinks.groupby("Beverage_category")["Calories"].mean().reset_index(name = "avg_cals")
```

# 6

```python
avg_cals = drinks["Calories"].mean()

high_cal = drinks.loc[drinks["Calories"] > avg_cals, "Beverage_prep"].reset_index()

print(drinks['Beverage_prep'].unique())
```

# 7

```python
coffee_drinks = drinks.loc[drinks["Beverage"] == "Coffee"]
coffee_drinks.loc[coffee_drinks["Calories"].idxmin()]
```

# 8

```python
espresso_drinks = drinks.loc[drinks["Beverage_category"] == "Classic Espresso Drinks"].copy()
espresso_drinks
```

# 9

```python
espresso_drinks.groupby("Beverage_prep")["Calories"].mean().reset_index()
```

# 10

```python
espresso_drinks.groupby("Beverage_prep")["Calories"].mean().reset_index().sort_values("Beverage_prep", ascending = False)
```
